```python
import sys, pyspark
print("Python:", sys.version)
print("PySpark:", pyspark.__version__)
```

```python
from pyspark.sql import SparkSession
spark = (SparkSession.builder
         .appName("sanity")
         .master("local[4]")
         .getOrCreate())
```

```python
import time
from pyspark import TaskContext
sc = spark.sparkContext
st = sc.statusTracker()

print("Master:", sc.master)                 # e.g., local[2]
print("defaultParallelism:", sc.defaultParallelism)
print("UI:", sc.uiWebUrl)
```

```python
def slow(x):
    time.sleep(5)
    return (TaskContext.get().partitionId(), x)

rdd = sc.parallelize(range(8), 8)
# Start a polling thread or just watch the UI
rdd.map(slow).count()  # you’ll see up to 2 active tasks at once on local[2]
```

```python
# Verify executor sees the same Python
sc.parallelize([0], 1).mapPartitions(
    lambda it: [(__import__("sys").version, __import__("sys").executable)]
).collect()

# Executors & memory map (confirms backend is up)
print(sc._jsc.sc().getExecutorMemoryStatus().toString())
```
